```python
%%time
# Load the data from the .data file into a Numpy array
data = pd.read_csv('nyc_energy_consumption.txt')

# Create a pandas DataFrame from the Numpy array
df = pd.DataFrame(data)

# Assign column names to the DataFrame
df.columns = ['timeStamp', 'demand', 'precip', 'temp']

# Save the DataFrame to a CSV file
df.to_csv('./Machine Learning/CSV Files/NYC_Energy_Consumption.csv', index=False)
df
```
